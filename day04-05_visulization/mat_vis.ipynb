```python
plt.figure(figsize=(6,3))
plt.title('GDP Over the Years', fontsize=15, color='#F87C63', fontweight=800)
plt.xlabel('------  Year ----->')
plt.ylabel('------ GDP ------>')

# Line Chart
plt.plot(year,gdp,
         linestyle='dotted', color='k',
         marker='*', markersize=10, markerfacecolor='#F87C63', markeredgecolor='purple'
        )
plt.show()
```

#### b. Scatter Plot

```python
prices = pd.read_csv("F:/offline _ds/00_datasets/house_prices.csv")
prices.head()
```

```python
prices.rename(
    columns={
        'Property #':'Property_No',
        'Area (ft.)':'Area'
    }, inplace=True
)
```

```python
prices.head()
```

```python
plt.figure(figsize=(6,3))
plt.title('House Prices')
plt.xlabel('------  Area (in Sq.Ft) ----->')
plt.ylabel('------ House Price ------>')

# Scatter Plot
plt.scatter(prices['Area'], prices['Price'],
            color='#F5276C', edgecolor='k', 
            alpha=0.7,
            s=70
           )
plt.show()
```

### Categorical vs Nr

```python
#### 1. Bar Chart
```

```python
cars =pd.read_csv("F:/offline _ds/00_datasets/bar_chart_data.csv")
cars
```

```python
plt.figure(figsize=(7,3))
plt.title('Used Cars')
plt.xlabel('------  Brand ----->')
plt.ylabel('------ No of Listings ------>')


# Bar Chart
ax = plt.bar(cars['Brand'], cars['Cars Listings'],
        color='tomato', edgecolor='k', alpha=0.7,
        
       )
plt.bar_label(ax)
plt.ylim(0,1000)
plt.xticks(fontsize=8, rotation=10)
plt.show()
```

```python
plt.figure(figsize=(7,3))
plt.title('Used Cars')
plt.xlabel('------  Brand ----->')
plt.ylabel('------ No of Listings ------>')

cars_ord = cars.sort_values(by='Cars Listings')
# Bar Chart
ax = plt.barh(cars_ord['Brand'], cars_ord['Cars Listings'],
        color='#7CCF35', edgecolor='k', alpha=0.7,
        
       )
plt.xlim(200,1000)
plt.bar_label(ax)
plt.xticks(fontsize=8)
plt.show()
```

#### 2. Pie Chart

```python
courses = ['Data Science','Analytics','Power BI','Tableau','Excel','MySql']
students = [130,227,280,189,67,98]
```

```python
# Pie Chart
plt.pie(students, labels=courses,
        autopct = '%0.1f%%',
        explode = [0,0,0,0,0.2,0],
        wedgeprops=dict(linewidth=1, edgecolor='darkgrey'),
        colors = ['#FFE2E2','#FFEDD4','#FFA1AD','#96F7E4','#DAB2FF','#90A1B9']
       )
plt.show()
```
