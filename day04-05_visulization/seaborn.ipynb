```python
cars =pd.read_csv("F:/offline _ds/00_datasets/bar_chart_data.csv")
cars
```

```python
sns.catplot(cars,x='Brand', y='Cars Listings',
            kind='bar',
            height=3,
            aspect=2,
            hue='Brand',
            palette='vlag',
            edgecolor='k'
           
)
plt.title('used cars listings')

plt.show()
```

### strip plot

```python
sns.catplot(tips,x='day',y='total_bill',
            kind='strip',
            palette='vlag',
            hue='sex'
)
plt.show()
```

```python
sns.catplot(tips,x='day',y='total_bill',
            kind='strip',
            palette='vlag',
            hue='time',
            height=3,
            aspect=1.5
)
plt.show()
```

```python
sns.catplot(tips,x='day',y='total_bill',
            kind='swarm',
            palette='vlag',
            hue='time',
            height=3,
            aspect=1.5
)
plt.show()
```

```python
sns.catplot(tips,x='day',y='total_bill',
            kind='box',
            palette='vlag',
            # hue='time',
            height=3,
            aspect=1.5
)
plt.show()
```

```python
sns.catplot(tips,x='day',y='total_bill',
            kind='violin',
            palette='mako',
            # hue='time',
            height=3,
            aspect=1.5
)
plt.show()
```

### relplot

```python
mpg=sns.load_dataset('mpg')
mpg.head()
```

```python
mpg.nunique()
```

```python
mpg['origin'].nunique()
```

```python
mpg['origin'].unique()
        
```

```python
sns.relplot(mpg, x='model_year',y='mpg',
            kind='line',
            # palette='mako',
            # hue='time',
            height=3,
            aspect=1.5
)
plt.show()
```

```python
sns.relplot(mpg, x='model_year',y='mpg',
            kind='line',
            palette='mako',
            height=3,
            aspect=1.5,
            errorbar=('ci',False),
            hue='origin',
            style='origin'
)
plt.show()
```

```python
sns.relplot(tips, x='total_bill',y='tip',
            kind='scatter',
            palette='mako',
            height=3,
            aspect=1.5,
            hue='sex',
            size='time',
            edgecolor='k'
)
plt.show()
```

```python
sns.displot(tips, x='total_bill',
            kind='kde',
            color='green',
            height=3,
            aspect=1.5,
            fill=True
           
)
plt.show()
```

```python
sns.displot(tips, x='total_bill',
            kind='hist',
            color='green',
            height=3,
            aspect=1.5,
            kde=True
           
)
plt.show()
```
