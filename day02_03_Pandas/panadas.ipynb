## data cleaning

### handling null values

```python
import numpy as np
```

```python
dnull = [ [1,2,3,'M',7],
         [np.nan,5,6,np.nan,7],
         [8,np.nan,9,'F',10],
         [np.nan,np.nan,np.nan,'M',11] 
        ]

dframe = pd.DataFrame(dnull, columns=['a','b','c','d','e'])
dframe
```

```python
dframe.info()
```

```python
dframe.isna()
```

```python
dframe.isna().sum()
```

```python
dframe['d'].mode()
```

```python
mode_d=dframe['d'].mode()[0]
mode_d
```

```python

avg_a = dframe['a'].mean()
avg_b = dframe['b'].mean()
avg_c = dframe['c'].mean()

avg_a ,avg_b , avg_c
```

```python
dframe
```

```python
# dframes
```

```python
dframe.fillna(
    {
        'a':avg_a ,
        'b':avg_b ,
        'c':avg_c,
        'd':mode_d
        
    },
    # inplace = true
    
)
```

```python
dframe.fillna(0)
```

```python
dframe.dropna(
    axis=0, #default
)
```

```python
dframe.dropna(
    axis=1, #default
)
```

```python
d1 = {'movieid':[1,2,3,4,5],
     'title':['3 Idiots','2 States','Leo','RRR','Pushpa']}
movies = pd.DataFrame(d1)


d2 = {
    'memid':[1,2,3,4],
    'name':['Tom','Tim','Jon','Jim'],
    'movieid':[1,np.nan,3,5]
}
members=pd.DataFrame(d2)
```

```python
movies
```

```python
members
```

```python
pd.merge(movies,members,on='movieid', how ='inner') # show all movies that have been rented and the members who have ranted
```

```python
pd.merge(movies,members,on='movieid', how ='left')
```

```python
pd.merge(movies,members,on='movieid', how ='right')
```

```python
pd.merge(movies,members,on='movieid', how ='outer')
```

### apply() - function on a series

```python
def myfxn(x):
 
    if x > 35000:
        res='expesive'
    else:
        res='affortable'
    return res
        
myfxn(30000) 
    
```

```python
u1['Expenses'].apply(myfxn)
```

```python
u1['Status']=u1['Expenses'].apply(myfxn)
u1
```
