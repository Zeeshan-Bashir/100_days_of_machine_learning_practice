### grading system<!--  -->

```python
marks=int(input('enter your marks'))
if marks >=80:
    print('distintion')
elif marks >=60 and marks <80 :
    print('grade A')
elif marks >=40 and marks <60 :
    print('grade b')
else:
    print('fail')
    # we can also use result = 'grade a'


      
```

### loops

```python
# for loop
# grading system usin fo loop
marks_list =[60,50,70,90,45,77]
for marks in marks_list:
    if marks >=80:
        res='distintion'
    elif marks >=60 and marks <80 :
        res='grade A'
    elif marks >=40 and marks <60 :
         res ='grade b'
    else:
        res='fail'
    print(marks,'=>',res)
    
    
    
```

### continue statement

```python
lst=[1,2,3,5,6,7,11,8,9]
for x in lst:
    if x==7:
        continue
    print(x)
```
