#### zip() function

```python
names = ['Tom','Jon','Pat']
marks = [78,89,90]
```

```python
zip(names, marks)
```

```python
for x in zip(names,marks):
    print(x)
```

```python
std_marks = dict(zip(names,marks))
std_marks
```

```python
std_marks['Jon']
```

```python
states = ['TN','Goa','AP','UP','MP','Bihar']
capitals = ['Chennai','Panaji','Amaravathi','Lucknow','Bhopal','Patna']
```

```python
dstates = dict(zip(states,capitals))
dstates
```
